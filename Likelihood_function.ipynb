```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
x = 5
X = 7
mu = np.linspace(0,10,100)
```

```python
Y = norm.pdf(x, mu, 1 )*norm.pdf(X, mu, 1 )
```

```python
plt.plot(mu, Y)
```

```python
x = np.random.normal(5, 1,10)
mu = np.linspace(0,10,100)
```

```python
ll = 0
for i in x:
   
    ll += np.log(norm.pdf(i, mu, 1))
```

```python
plt.plot(mu, ll)
```

```python
ll = 1
for i in x:
   
    ll = ll*(norm.pdf(i, mu, 1))
```

```python
plt.plot(mu, ll)
```
