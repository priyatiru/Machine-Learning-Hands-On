```python
import matplotlib.pyplot as plt
```

```python
eta = 1

def error_fun(data, w):
	a = w[0]
	for i in range(len(data)-1):
		a += w[i + 1] * data[i]
	if(a>=0.0):
		return 1.0
	else:
		return -1.0
```

```python
def perceptron(dataset, itration):
	w = [0.0 for i in range(len(dataset[0]))]
	for i in range(itration):
		print(str(i)+" ",w) 	
		total_error = 0
		for data in dataset:
			pred_class = error_fun(data, w)
			error = pred_class * data[-1]
			print("data :  " ,data )
			print("class " ,data[-1] )
			print("pred_class " ,pred_class )
			if(error<0.0):
				total_error +=1
				w[0] = w[0] + eta * data[-1]
				for t in range(len(data)-1):
					w[t + 1] = w[t + 1] + eta * data[-1] * data[t]
			print(w) 
		print("total error  : " , total_error)
		print("\n")
		
	return w
```

```python
dataset = [[0.26,0.58,-1], [0.1,1.5,1]  ,[0.45,0.15,-1], [0.35,0.95,1] ,[0.60,0.30,-1],[0.70,0.65,1] ,[0.08,0.72,-1] , [0.92,0.45,1]]

itration = 10
w = perceptron(dataset,itration)
print(w)
```

```python
for data in dataset:
	if(data[2] == 1):
		plt.scatter(data[0],data[1] , c='r')
	else:
		plt.scatter(data[0],data[1] , c='b')
x=[-w[0]/w[1] , 0.0]
y=[0.0,-w[0]/w[2]]

plt.plot(y,x)
plt.show()
```
